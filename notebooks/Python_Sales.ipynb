```python
# Prophet forecasting example (Colab-ready)
import pandas as pd
import matplotlib.pyplot as plt

# --- Prepare your data (assumes df already loaded in session) ---
# If you don't have df loaded, load it first, e.g.:
# df = pd.read_csv('/path/to/superstore.csv')

# Ensure Order Date is datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True, errors='coerce')

# Group to daily sales
forecast_df = df.groupby('Order Date', as_index=False)['Sales'].sum()
forecast_df.columns = ['ds', 'y']

# Ensure ds is datetime and sorted
forecast_df['ds'] = pd.to_datetime(forecast_df['ds'])
forecast_df = forecast_df.sort_values('ds').reset_index(drop=True)

# Optional: set frequency and fill missing dates (Prophet can handle irregular dates but fill is often helpful)
forecast_df = forecast_df.set_index('ds').asfreq('D').fillna(method='ffill').reset_index()

# --- Fit Prophet model ---
from prophet import Prophet

m = Prophet()
m.fit(forecast_df)

# Make future dataframe and predict
future = m.make_future_dataframe(periods=90)   # 90 days into future
forecast = m.predict(future)

# Plot
fig1 = m.plot(forecast)
plt.title("Sales Forecast for Next 3 Months (Prophet)")
plt.show()

# Optional: plot components
fig2 = m.plot_components(forecast)
plt.show()
```

```python
# 1) Inspect df
import pandas as pd
import numpy as np

print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

print("\nData types:")
print(df.dtypes)

print("\nSample rows:")
display(df.head())

print("\nMissing values per column:")
print(df.isnull().sum())

# Show unique counts for categorical columns we care about
for col in ['Region','Category','Sub-Category','Ship Mode','Segment']:
    if col in df.columns:
        print(f"\nUnique in {col}:", df[col].nunique())
        print(df[col].value_counts().head(10))
```

```python
# 2) Clean column names and duplicates
df.columns = df.columns.str.strip()  # remove stray spaces
# Trim whitespace in string columns
for c in df.select_dtypes(include=['object']).columns:
    df[c] = df[c].astype(str).str.strip()

# Drop exact duplicate rows
before = df.shape[0]
df.drop_duplicates(inplace=True)
after = df.shape[0]
print(f"Dropped {before-after} duplicate rows.")
```

```python
# 3) Missing values details
missing = df.isnull().sum().sort_values(ascending=False)
missing_pct = (missing / len(df) * 100).round(2)
missing_report = pd.concat([missing, missing_pct], axis=1)
missing_report.columns = ['missing_count','missing_pct']
display(missing_report.head(30))

# Quick actions (do not auto-change anything without review)
# Examples (uncomment to run after review):
# df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
# df['Postal Code'] = df['Postal Code'].astype('str')
```

```python
# 4) Date parsing + features
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True, errors='coerce')
df['Ship Date']  = pd.to_datetime(df['Ship Date'], dayfirst=True, errors='coerce')

# Feature extraction
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Day'] = df['Order Date'].dt.day
df['Weekday'] = df['Order Date'].dt.day_name()
df['Order_Month_Year'] = df['Order Date'].dt.to_period('M').astype(str)

print("Converted Order Date dtype:", df['Order Date'].dtype)
display(df[['Order Date','Year','Month','Day','Weekday']].head())
```

```python
# 5) Ensure numeric columns are numeric
if 'Sales' in df.columns:
    df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')

# If you had Profit/Quantity/Discount, convert similarly:
for col in ['Profit','Quantity','Discount']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

print(df[['Sales']].describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

# A: Top 10 categories by total sales
cat_sales = df.groupby('Category', observed=True)['Sales'].sum().sort_values(ascending=False)
top_cat = cat_sales.head(10)
plt.figure(figsize=(10,5))
top_cat.plot(kind='bar')
plt.title("Top 10 Categories by Total Sales")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# B: Region-wise total sales
reg_sales = df.groupby('Region', observed=True)['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(7,4))
sns.barplot(x=reg_sales.index, y=reg_sales.values)
plt.title("Total Sales by Region")
plt.ylabel("Total Sales")
plt.xlabel("Region")
plt.tight_layout()
plt.show()

# C: Monthly sales trend (time series)
ts = df.groupby(pd.Grouper(key='Order Date', freq='M'))['Sales'].sum().reset_index()
plt.figure(figsize=(12,5))
sns.lineplot(data=ts, x='Order Date', y='Sales', marker='o')
plt.title("Monthly Sales Trend")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 7) Daily total series for forecasting
forecast_df = df.groupby('Order Date', as_index=False)['Sales'].sum().rename(columns={'Order Date':'ds','Sales':'y'})
# ensure ds sorted and datetime
forecast_df['ds'] = pd.to_datetime(forecast_df['ds'])
forecast_df = forecast_df.sort_values('ds').reset_index(drop=True)

# Ensure daily frequency and handle small gaps (choose fill method)
forecast_ts = forecast_df.set_index('ds').asfreq('D')   # introduces NaT for missing dates
# Option: forward fill small gaps; or interpolate, or leave as-is for Prophet
forecast_ts['y'] = forecast_ts['y'].fillna(method='ffill')  # or .interpolate()

forecast_ts = forecast_ts.reset_index()
display(forecast_ts.head(10))
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import matplotlib.pyplot as plt

# Prepare series
ts = forecast_ts.set_index('ds')['y']
# If there are still NaNs after fill, fill again
ts = ts.fillna(method='ffill')

# Fit Holt-Winters with weekly seasonality (change seasonal_periods as needed)
hw = ExponentialSmoothing(ts, trend='add', seasonal='add', seasonal_periods=7).fit()
pred = hw.forecast(90)

plt.figure(figsize=(12,5))
plt.plot(ts.index, ts.values, label='History')
plt.plot(pred.index, pred.values, label='Forecast (Holt-Winters)')
plt.title("Sales Forecast (Next 90 days) - Holt-Winters")
plt.legend()
plt.tight_layout()
plt.show()
```
